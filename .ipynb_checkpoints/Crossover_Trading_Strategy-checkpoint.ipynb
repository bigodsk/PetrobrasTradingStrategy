# Importing Dataset

```python
# download PETR4.SA 2022 historical data using the daily time frame 
# you can select any others datasets from www.finance.yahoo.com
df_petr = yf.download('PETR4.SA', start='2022-01-01', end='2022-12-31')
df_petr
```

# Data Information
* Open: the opening value
* High: maximum value in this date
* Low: minimum value in this date
* Close: close price adjusted for splits
* Adj Close: price adjusted for splits and dividend and/or capital gain distributions
* Volume: number of a stock’s shares that are traded on a stock exchange in a day

```python
# you can choose this following values: ema_fast, ema_slow

# exponencial moving average (fast): EMA_0
ema_fast = 5
# exponencial moving average (slow): EMA_1 
ema_slow = 10

# creating the exponential moving averages
EMA_0 = df_petr['Close'].ewm(span=ema_fast, adjust=False).mean()
EMA_1 = df_petr['Close'].ewm(span=ema_slow, adjust=False).mean()
```

```python
# visualizing PETR4.SA graph from the chosen period with the Exponential Moving Averages
plt.figure(figsize=(16,8))
plt.plot(df_petr['Close'], label='Petrobras', alpha=.7)
plt.plot(EMA_0, label=f'EMA - {ema_fast} periods')
plt.plot(EMA_1, label=f'EMA - {ema_slow} periods')

plt.legend()
```

```python
# creating cols with the moving averages
df_petr['EMA_0'] = EMA_0
df_petr['EMA_1'] = EMA_1
```

```python
df_petr
```

```python
# creating buy/sel indicators according to exponential moving average crossover
def crossover(df):
    signal_buy = []
    signal_sell = []
    flag = 0
    
    for i in range(0, len(df)):
        if df['EMA_0'][i] > df['EMA_1'][i]:
            if flag != 1:
                signal_buy.append(df['Close'][i])
                signal_sell.append(np.nan)
                flag = 1
            else:
                signal_buy.append(np.nan)
                signal_sell.append(np.nan)
        elif df['EMA_1'][i] > df['EMA_0'][i]:
            if flag != 2:
                signal_buy.append(np.nan)
                signal_sell.append(df['Close'][i])
                flag = 2
            else:
                signal_buy.append(np.nan)
                signal_sell.append(np.nan)
        else:
            signal_buy.append(np.nan)
            signal_sell.append(np.nan)
            
    return (signal_buy, signal_sell)
```

```python
len(crossover(df_petr))
```

```python
df_petr['Signal_Buy'] = crossover(df_petr)[0]
df_petr['Signal_Sell'] = crossover(df_petr)[1]
```

```python
df_petr
```

```python
plt.figure(figsize=(16,8))
plt.plot(df_petr['Close'], label='Petrobras', alpha=.7)
plt.plot(df_petr['EMA_0'], label=f'EMA - {ema_fast} periods')
plt.plot(df_petr['EMA_1'], label=f'EMA - {ema_slow} periods')
plt.plot(df_petr['Signal_Buy'], marker='^', color='green', label='Buy Signal')
plt.plot(df_petr['Signal_Sell'], marker='v', color='red', label='Sell Signal')

plt.title('Moving Average Crossover Strategy', fontsize=18)
plt.ylabel('Capital (R$)', fontsize=14)

plt.legend()
```
